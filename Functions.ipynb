```python
# 3. What are the different ways to define and call a function in Python?

'''There are 2 ways to call a function: -'''

# 1) Using inbuilt functions- We can use in-built functions like ceil, floor, etc to define a function.
from math import floor
a = floor(3.555)
print(a)

# 2) Using user-defined functions- Like using a keyword def, we can define a function.

def add_1(a,b):   #Here, add_1 is an user-defined function
    return a+b
print(add_1(2,3))

'''There are 4 ways to call a function: -'''

# 3) Using default, positional, keyword, and variable length arguments- These arguments can be passed to call the function as you want.

def add_1(a,b):   #Here, a & b are positional arguments. Here the order of passing the arguments matters.
    return a+b
print(add_1(2,3))

def add_2(a,b=3):   #Here, b is a default arguments i.e. even if we do not pass the other variable (b) , it will take b=4 as default argument.
    return a+b
print(add_2(2))

def add_3(a,b):   #Here, b=2,a=3 are keyword arguments.Here the developer itself assigns the order of passing the arguments.
    return a+b
print(add_3(b=2,a=3))

def sum(*args):     #variable length argument- when you do not know the no of arguments to be passed.
  s =0
  for i in args:
    s=s+i
  return s
print(sum(1,2,3))


```

```python
# 6. Explain the concept of generators in Python and how they are defined.

'''Generators are a simple way to create iterators using functions and the yield keyword instead of returning values.
They are defined using the keyword yield.'''

def square_number_generator(n):
    for i in range (n):
        yield i**2
print(square_number_generator(10))
gen=square_number_generator(10)
print(next(gen))
print(next(gen))
print(next(gen))


```

```python
# 7. What are the advantages of using generators over regular functions?

'''Advantages of generators over regular functions: -
Efficient Memory Usage: Generators do not store the entire sequence in memory, making them suitable for large datasets or infinite sequences.
Simpler Syntax: Using yield simplifies the creation of iterators, making the code more concise and readable.'''

#Regular function returns the results in one go, hence taking longer execution time.
def square_number(n):
    result=[]
    for i in range (n):
        result.append(i**2)
    return result
print(square_number(10))

#Generators shorten this execution time by returning element one by one. Also makes the code more readable
def square_number_generator(n):
    for i in range (n):
        yield i**2
print(square_number_generator(10))
gen=square_number_generator(10)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a,b=b,a+b
print (fib(100000000))  #When we work with large numbers, this eases the execution time by displaying result one by one.
f=fib(1000000000)
print(next(f))
print(next(f))
print(next(f))
print(next(f))


```
